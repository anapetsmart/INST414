```python
import pandas as pd
import json
```

```python
eye_df =  pd.read_csv("full_df.csv")
eye_df["diagnostic"] = eye_df[["Left-Diagnostic Keywords", "Right-Diagnostic Keywords"]].apply(lambda x: ', '.join(x[x.notnull()]), axis=1)
eye_df= eye_df.drop(columns=['Left-Fundus', 'Right-Fundus', 'Left-Diagnostic Keywords', 'Right-Diagnostic Keywords', 'N', 'D', 'G', 'C', 'A', 'H', 'M', 'O', 'filepath', 'labels', 'target', 'filename'])
eye_json = eye_df.to_json(orient='records')
eye_json = json.loads(eye_json)
eye_json
eye_disease = []
eye_disease_map = {}
eye_disease_map['ID'] = {}
for line in eye_json:
    line['diagnostic'] = line['diagnostic'].replace('，', ',')
    line['diagnostic'] = [i.strip() for i  in line['diagnostic'].split(',')] 
#eye_json
for line in eye_json:
    for i in line['diagnostic']:
        if i not in eye_disease:
            eye_disease.append(i)
    eye_disease_map[line['ID']] = {}
    for i in eye_disease:
        #count = 0 
        if i in line['diagnostic']:
        #for i in line['diagnostic']: 
        #        count = count + 1 
            eye_disease_map[line['ID']][i] = line['diagnostic'].count(i)
        else:
            eye_disease_map[line['ID']][i] = 0
            count = 0
#eye_disease_map
eye_disease_df = pd.DataFrame(eye_disease_map)
#eye_disease_df = eye_disease_df.swapaxes('index', 'columns')
eye_disease_df = eye_disease_df.fillna(0)
# eye_disease_df = eye_disease_df.sum(axis=1)
```

```python
normed_rows = []
L1 = eye_disease_df.sum(axis = 1)
for index, row in eye_disease_df.iterrows():
    normed_rows.append(row/L1[index])
normed_eye_disease = pd.DataFrame(normed_rows)
normed_eye_disease.sum(axis=1)
normed_eye_disease
```

```python
from sklearn.metrics import DistanceMetric
from sklearn.metrics.pairwise import cosine_distances
import scipy.spatial.distance
dist = cosine_distances
eye_disease_df
```

```python
target_patient_id = 'normal fundus'
target_patient_ratings = normed_eye_disease.loc[target_patient_id]
distances = scipy.spatial.distance.cdist(normed_eye_disease, [target_patient_ratings], metric = 'euclidean')[:,0]
query_distances =list(zip(normed_eye_disease.index, distances))
for similar_patient_id in sorted(query_distances, key=lambda x:x[1], reverse = False)[:11]:
      print(similar_patient_id)
         
#, eye_disease_map[similar_patient_id], eye_disease_df[similar_patient_id].sum())

```
